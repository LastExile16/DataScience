## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
raw_data = pd.read_csv('Countries-exercise.csv')
raw_data.describe()
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(raw_data['Longitude'], raw_data['Latitude'])
plt.xlabel('Logitude')
plt.ylabel('Latitude')
plt.xlim(-200,200)
plt.ylim(-90, 90)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
data = raw_data.copy()
data = data.iloc[:, 1:3]
data.head()
```

## Clustering

Assume there are only two clusters. 

```python
kmeans = KMeans(8)
kmeans.fit(data)
```

```python
predicted_clusters = kmeans.fit_predict(data)
```

### Clustering Resutls

```python
data_clusters = data.copy()
data_clusters['predicted_clusters'] = predicted_clusters
```

```python
plt.scatter(data_clusters['Longitude'], data_clusters['Latitude'], c=data_clusters['predicted_clusters'], cmap='rainbow')
```
